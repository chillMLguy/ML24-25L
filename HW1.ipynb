**1.Wczytujemy dane i biblioteki**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
url = "https://raw.githubusercontent.com/SzymonNowakowski/Machine-Learning-2024/master/PCA_three_clusters_data.csv"
data = pd.read_csv(url)

X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

print(data.head())
print(data.shape)
```

**2.PCA**


```python
#scalling data
scaler = StandardScaler()
X = scaler.fit_transform(X)

#pca
pca = PCA(n_components = 2)
X_new = pca.fit_transform(X)

#plotting
plt.figure(figsize=(8, 8))
for i, target_name in enumerate([0,1,2]):
    plt.scatter(X_new[y==i, 0], X_new[y==i, 1], label=target_name, alpha=0.7)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA of dataset")
plt.legend(title="class")
plt.grid(True)
plt.show()
```

**3.Porównanie**

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

for i, target_name in enumerate([0,1,2]):
    axes[0,0].scatter(X_new[y==i, 0], X_new[y==i, 1], label=target_name, alpha=0.7)
axes[0,0].set_xlabel("Principal Component 1")
axes[0,0].set_ylabel("Principal Component 2")
axes[0,0].set_title("PCA of dataset")
axes[0,0].grid(True)

for i, target_name in enumerate([0,1,2]):
    axes[0,1].scatter(X[y==i, 0], X[y==i, 1], label=target_name, alpha=0.7)

axes[0,1].set_xlabel("x")
axes[0,1].set_ylabel("y")
axes[0,1].set_title("x vs. y")
axes[0,1].grid(True)

for i, target_name in enumerate([0,1,2]):
    axes[1,0].scatter(X[y==i, 0], X[y==i, 2], label=target_name, alpha=0.7)

axes[1,0].set_xlabel("x")
axes[1,0].set_ylabel("z")
axes[1,0].set_title("x vs. z")
axes[1,0].grid(True)

for i, target_name in enumerate([0,1,2]):
    axes[1,1].scatter(X[y==i, 1], X[y==i, 2], label=target_name, alpha=0.7)

axes[1,1].set_xlabel("y")
axes[1,1].set_ylabel("z")
axes[1,1].set_title("y vs. z")
axes[1,1].grid(True)


plt.show()
```
